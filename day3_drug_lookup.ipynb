A dictionary — key: value pairs

```python
# A dictionary — key: value pairs
drug = {
    "name" : "Metformin",
    "category" : "Antidiabetic",
    "use" : "Managing type 2 diabetes",
    "side_effects" : "Nausea, stomach upset"
}

# Access a value using its key
print (drug["name"])
print (drug["category"])
print (drug["use"])
print (drug["side_effects"])
```

Dictionary of dictionaries — like a mini database

```python
# Dictionary of dictionaries — like a mini database

pharma_db = {
    "Metformin" : {
        "category": "Antidiabetic",
        "use": "Managing type 2 diabetes",
        "side_effects": "Nausea, stomach upset"
    },

    "Amoxicillin" : {
        "category" : "Antibiotic",
        "use" : "Treating bacterial infections",
        "side_effects" : "Rash, diarrhea"
    },

    "Omeprazole" : {
        "category" : "PPI",
        "use" : "Reducing stomach acid",
        "side_effects" : "Headache, nausea"
    }
}

print (pharma_db["Metformin"])
print (pharma_db["Amoxicillin"]["use"])
```

Drug Lookup Tool — Day 3 Project

```python
# Drug Lookup Tool — Day 3 Project

pharma_db = {
    "Metformin": {
        "category": "Antidiabetic",
        "use": "Managing type 2 diabetes",
        "side_effects": "Nausea, stomach upset"
    },
    "Amoxicillin": {
        "category": "Antibiotic",
        "use": "Treating bacterial infections",
        "side_effects": "Rash, diarrhea"
    },
    "Omeprazole": {
        "category": "PPI",
        "use": "Reducing stomach acid",
        "side_effects": "Headache, nausea"
    },
    "Paracetamol": {
        "category": "Analgesic",
        "use": "Pain relief and fever reduction",
        "side_effects": "Rare at normal doses"
    },
    "Atorvastatin": {
        "category": "Statin",
        "use": "Lowering cholesterol levels",
        "side_effects": "Muscle pain, liver issues"
    }
}

def lookup_drug(drug_name) :
  if drug_name in pharma_db :
    drug = pharma_db[drug_name]
    print(f"\ndrug : {drug_name}")
    print(f"category : {drug['category']}")
    print(f"use : {drug['use']}")
    print(f"side_effects : {drug['side_effects']}")
    print("_" * 40)
  else:
    print(f"\n '{drug_name}' not found in database.")

 # Test the lookup

lookup_drug("Metformin")
lookup_drug("Paracetamol")
lookup_drug("Ibuprofen")
```
