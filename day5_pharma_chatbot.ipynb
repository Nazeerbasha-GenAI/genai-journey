```python
from groq import Groq
from google.colab import userdata

api_key = userdata.get('GROQ_API_KEY')
client = Groq(api_key=api_key)

# Conversation history
conversation_history = []

# System prompt
system_prompt = """You are a helpful pharmaceutical expert assistant
working with the Roche pharma ops team. Answer questions about
drugs, clinical trials, SOPs, and pharma operations clearly and
concisely. If a question is not pharma related, politely decline."""

def chat(user_input):
    # Add user message to history
    conversation_history.append({
        "role": "user",
        "content": user_input
    })

    # Send to AI
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {"role": "system", "content": system_prompt}
        ] + conversation_history
    )

    # Get reply
    ai_reply = response.choices[0].message.content

    # Add reply to history
    conversation_history.append({
        "role": "assistant",
        "content": ai_reply
    })

    print(f"You: {user_input}")
    print(f"\nChatbot: {ai_reply}")
    print("-" * 40)
```

```python
chat("What is Metformin used for?")
```

```python
chat("What are its side effects?")
```

```python
chat("Is it safe for elderly patients?")
```
