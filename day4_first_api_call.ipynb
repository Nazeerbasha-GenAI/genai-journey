```python
# Step 2 — Import the tools we need
from google.colab import userdata   # tool to read our secret key safely
from groq import Groq               # tool to talk to Groq's AI

# Step 3 — Load your API key from Colab Secrets
# This finds your key by its name — never exposes the full key in code
api_key = userdata.get('GROQ_API_KEY')

# Step 4 — Create a connection to Groq using your key
# Think of this like logging into Groq's servers
client = Groq(api_key=api_key)

# Step 5 — Send a message to the AI and store the reply
# model   = which AI brain to use
# messages = the conversation — role tells who is speaking
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {
            "role": "user",         # "user" means this is your message
            "content": "What is pharmacovigilance? Answer in 2 sentences."
        }
    ]
)

# Step 6 — Print the AI's reply
# The answer is buried inside the response object — this digs it out
print(response.choices[0].message.content)
```

This is the trickiest line — let's break it down piece by piece:
response                  → the full envelope that came back from Groq
response.choices          → open the envelope — list of possible replies
response.choices[0]       → take the first reply (usually only one)
response.choices[0].message        → the actual message inside
response.choices[0].message.content  → the text of that message
Think of it like a Russian nesting doll — you keep opening layers until you get to the actual answer inside.
