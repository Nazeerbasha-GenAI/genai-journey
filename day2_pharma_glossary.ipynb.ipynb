```python
# A function is a reusable block of code
# def means "define a new function"

def describe_drug(name, use):
  sentence = f"{name} is used for {use}"
  print (sentence)

# Now CALL the function with different drugs

describe_drug("Metformin", "managing type 2 diabetes")
describe_drug("Atorvastatin", "lowering cholesterol")
describe_drug("Paracetamol", "pain relief and fever reduction")
```

```python
# Pharma Drug Glossary — Day 2 Project

def describe_drug(name, category, use):
  print(f"\nDrug : {name}")
  print(f"category : {category}")
  print(f"use : {use}")
  print("_" * 35)

# Call the function for 5 drugs

describe_drug("Paracetamol", "Analgesic", "Pain relief and fever reduction")
describe_drug("Metformin", "Antidiabetic", "Managing type 2 diabetes")
describe_drug("Atorvastatin", "Statin", "Lowering cholesterol levels")
describe_drug("Amoxicillin", "Antibiotic", "Treating bacterial infections")
describe_drug("Omeprazole", "PPI", "Reducing stomach acid")
```
